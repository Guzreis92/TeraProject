```python
#Importando bibliotecas para rodar o apriori:
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Café da manhã:
frequent_itemsets_cafe = apriori(compras_cafe, min_support=0.1,use_colnames=True)
rules_cafe = association_rules(frequent_itemsets_cafe, metric="lift", min_threshold=1)

#Almoço:
frequent_itemsets_almoco = apriori(compras_almoco, min_support=0.1,use_colnames=True)
rules_almoco = association_rules(frequent_itemsets_almoco, metric="lift", min_threshold=1)

#Jantar:
frequent_itemsets_jantar = apriori(compras_jantar, min_support=0.1,use_colnames=True)
rules_jantar = association_rules(frequent_itemsets_jantar, metric="lift", min_threshold=1)

#Lanche:
frequent_itemsets_lanche = apriori(compras_lanche, min_support=0.1,use_colnames=True)
rules_lanche = association_rules(frequent_itemsets_lanche, metric="lift", min_threshold=1)
```
